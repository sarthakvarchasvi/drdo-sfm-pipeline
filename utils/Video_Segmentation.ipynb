```python
import cv2
import os

# === INPUT ===
video_path = "Datasets/Malta-City/video.webm"
output_folder = "Datasets/Malta-City/Video_Frames"
frame_skip = 40

# === SETUP ===
os.makedirs(output_folder, exist_ok=True)
cap = cv2.VideoCapture(video_path)
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
count = 0
saved = 0

# === EXTRACTION LOOP ===
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    if count % frame_skip == 0:
        filename = os.path.join(output_folder, f"frame_{saved:04d}.jpg")
        cv2.imwrite(filename, frame)
        saved += 1
    count += 1

cap.release()
print(f"Extracted {saved} frames from {video_path} into {output_folder}/")
```
